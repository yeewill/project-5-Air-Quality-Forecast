```python
import plaidml.keras
plaidml.keras.install_backend()

import pickle
from sklearn.preprocessing import MinMaxScaler, StandardScaler
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.metrics import mean_squared_error
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LSTM
from keras.layers import SimpleRNN
from keras.layers import Dropout
from keras.layers import Activation
from keras.callbacks import EarlyStopping



pd.set_option('display.max_columns', 500)
```

```python
dataset= pickle.load(open('ts_with_weather_sf.p', 'rb'))
```

```python
# convert series to supervised learning
def series_to_supervised(data, n_in=1, n_out=1, dropnan=True):
    n_vars = 1 if type(data) is list else data.shape[1]
    df = pd.DataFrame(data)
    cols, names = list(), list()
    # input sequence (t-n, ... t-1)
    for i in range(n_in, 0, -1):
        cols.append(df.shift(i))
        names += [('var%d(t-%d)' % (j+1, i)) for j in range(n_vars)]
    # forecast sequence (t, t+1, ... t+n)
    for i in range(0, n_out):
        cols.append(df.shift(-i))
        if i == 0:
            names += [('var%d(t)' % (j+1)) for j in range(n_vars)]
        else:
            names += [('var%d(t+%d)' % (j+1, i)) for j in range(n_vars)]
    # put it all together
    agg = pd.concat(cols, axis=1)
    agg.columns = names
    # drop rows with NaN values
    if dropnan:
        agg.dropna(inplace=True)
    return agg
 
```

```python
# load dataset
values = dataset.values
# integer encode direction
# ensure all data is float
values = values.astype('float32')
# normalize features
scaler = MinMaxScaler()
scaled = scaler.fit_transform(values)
# frame as supervised learning
reframed = series_to_supervised(scaled, 7, 1)
# drop columns we don't want to predict

reframed.head()
```

```python
y = reframed.iloc[:,-11].values
X = reframed.drop('var1(t)', axis =1).values
# split into train and test sets

n_train_days = 2300
n_test_days=360
train_X = X[:n_train_days,:]
train_y = y[:n_train_days]

val_X= X[n_train_days:-n_test_days,]
val_y= y[n_train_days:-n_test_days]

test_X = X[-n_test_days:,:]
test_y = y[-n_test_days:]



# reshape input to be 3D [samples, timesteps, features]
train_X = train_X.reshape((train_X.shape[0], 1, train_X.shape[1]))
val_X = val_X.reshape((val_X.shape[0], 1, val_X.shape[1]))
test_X = test_X.reshape((test_X.shape[0], 1, test_X.shape[1]))
print(train_X.shape, train_y.shape,val_X.shape, val_y.shape ,test_X.shape, test_y.shape)
```

```python
##### # design network
model = Sequential()
model.add(LSTM(1200, input_shape=(train_X.shape[1], train_X.shape[2]), dropout=0.15, recurrent_dropout=0.15))

model.add(Activation("relu"))
model.add(Dense(2000))
model.add(Dropout(.01))
model.add(Activation("relu"))

# model.add(Activation("tanh"))
# model.add(Dense(200))
# model.add(Dropout(.2))
# model.add(Dense(200))
# model.add(Dropout(.2))
# model.add(Maxpooling())

model.add(Dense(1))
model.compile(loss='mae', optimizer='adam')

# fit network
history = model.fit(train_X, train_y, epochs=100, batch_size=20, validation_data=(val_X, val_y) ,verbose=2, shuffle=False)
# plot history


plt.plot(history.history['loss'], label='train')
plt.plot(history.history['val_loss'], label='val')
plt.legend()
plt.show()
```

```python
simple_rnn = Sequential()
simple_rnn.add(SimpleRNN(140, input_shape=(train_X.shape[1], train_X.shape[2]),recurrent_dropout=.2))
simple_rnn.add(Dense(1))
simple_rnn.compile(loss='mae', optimizer='adam')
               
history = simple_rnn.fit(train_X, train_y, epochs=100, batch_size=30, validation_data=(val_X, val_y), verbose=2, shuffle=False)
# plot history
plt.plot(history.history['loss'], label='train')
plt.plot(history.history['val_loss'], label='val')
plt.legend()
plt.show()
```

```python

# design network
stacked_model = Sequential()
stacked_model.add(LSTM(1200, input_shape=(train_X.shape[1], train_X.shape[2]), dropout=0.2, recurrent_dropout=0.4, 
                       return_sequences=True))
stacked_model.add(LSTM(1000, input_shape=(train_X.shape[1], train_X.shape[2]), dropout=0.3, recurrent_dropout=0.3 ))
stacked_model.add(Dense(200))
stacked_model.add(Dropout(.3))
stacked_model.add(Dense(1))

stacked_model.compile(loss='mae', optimizer='adam')
# fit network
history = stacked_model.fit(train_X, train_y, epochs=100, batch_size=20, validation_data=(val_X, val_y), verbose=2, shuffle=False)
# plot history
plt.plot(history.history['loss'], label='train')
plt.plot(history.history['val_loss'], label='test')
plt.legend()
plt.show()
```

```python
stacked_model.predict(test_X)
```

```python
preds = []
preds_array = stacked_model.predict(test_X)
for i in range(len(preds_array)):
    preds.append(preds_array[i][0])
```
